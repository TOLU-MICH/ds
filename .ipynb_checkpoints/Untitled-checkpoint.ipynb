```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the 'naturalearth_lowres' dataset from the specified URL
url = "https://www.naturalearthdata.com/downloads/110m-cultural-vectors/"
world=gpd.read_file('110m_cultural/ne_110m_admin_0_boundary_lines_land.shp')

# Plot the world map with population estimates
world.plot(column='pop_est', cmap='OrRd', legend=True)
plt.title('World Population Estimates')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
