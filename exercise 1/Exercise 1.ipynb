# Task 1

```python
# Required Libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the DJIA data
Djia_data = sns.load_dataset('dowjones')

# Load the DJIA data (assuming you've already uploaded the DowJones.csv file)
df = pd.read_csv('djia.csv')

# Calculate summary statistics
mean_value = df['Price'].mean()
median_value = df['Price'].median()
std_value = df['Price'].std()
min_value = df['Price'].min()
max_value = df['Price'].max()

# Print the results
print(f"Mean DJIA value: {mean_value:.2f}")
print(f"Median DJIA value: {median_value:.2f}")
print(f"Standard deviation of DJIA: {std_value:.2f}")
print(f"Minimum DJIA value: {min_value:.2f}")
print(f"Maximum DJIA value: {max_value:.2f}")
```




# Task 2

```python
# Create a box plot
plt.figure(figsize=(10, 7))
plt.boxplot(df['Price'])
plt.title('Distribution of DJIA Prices')
plt.xlabel('DJIA')
plt.ylabel('Price (USD)')
plt.show()
```

# Task 3

```python
# Create a violin plot
plt.figure(figsize=(10, 7))
sns.violinplot(x=df['Date'], y=df['Price'])
plt.title('Distribution of DJIA Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

# Task 4

```python
# Create a line chart
plt.figure(figsize=(10, 7))
plt.plot(df['Date'], df['Price'], marker='o', linestyle='-', color='b')
plt.title('DJIA Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the DJIA data (assuming you've already uploaded the DowJones.csv file)
df = pd.read_csv('djia.csv')

# Generate x-axis data (indices)
x = np.arange(len(df))

# Calculate coefficients for the line of best fit
a, b = np.polyfit(x, df['Price'], 1)

# Create the line of best fit
line_of_best_fit = np.poly1d([a, b])

# Plot the scatter plot of DJIA data
plt.figure(figsize=(10, 7))
plt.scatter(x, df['Price'], label='DJIA Data', color='b')

# Plot the line of best fit
plt.plot(x, line_of_best_fit(x), label='Line of Best Fit', color='r', linestyle='--')

plt.title('Line of Best Fit for DJIA Prices')
plt.xlabel('Index')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
