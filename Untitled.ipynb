```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the 'world' dataset (assuming you've already loaded it)
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a new figure
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the administrative boundaries of countries
world.boundary.plot(ax=ax, color='black', linewidth=0.5)

# Load a dataset containing major cities (you can replace this with your own data)
# For example, using the 'naturalearth_cities' dataset
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))

# Plot major cities on the same map
cities.plot(ax=ax, marker='o', color='red', markersize=5)

# Add labels for major cities (optional)
for x, y, label in zip(cities.geometry.x, cities.geometry.y, cities['name']):
    ax.text(x, y, label, fontsize=4, ha='center', va='center')

plt.title('Administrative Boundaries and Major Cities')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import geopandas as gp
import matplotlib.pyplot as plt

# Load the 'world' dataset
world = gp.read_file(gp.datasets.get_path('naturalearth_lowres'))

# Plot the world map with labels
ax = world.plot(column='pop_est', cmap='OrRd', legend=True)
plt.title('World Population Estimates')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add country names as labels
for x, y, label in zip(world.geometry.centroid.x, world.geometry.centroid.y, world['name']):
    ax.text(x, y, label, fontsize=0.4, ha='center', va='center')

# Add continent details (optional)
continents = world.dissolve(by='continent')
for x, y, label in zip(continents.geometry.centroid.x, continents.geometry.centroid.y, continents.index):
    ax.text(x, y, label, fontsize=1, ha='center', va='center', fontweight='bold')

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the 'world' dataset (assuming you've already loaded it)
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Remove Antarctica
world = world[world['name'] != 'Antarctica']

# Calculate GDP per capita
world['GDP_per_capita'] = world['gdp_md_est'] / world['pop_est']

# Define custom color thresholds
gdp_min = world['GDP_per_capita'].min()
gdp_max = world['GDP_per_capita'].max()
threshold_low = gdp_min + 0.2 * (gdp_max - gdp_min)
threshold_high = gdp_min + 0.8 * (gdp_max - gdp_min)

# Create a choropleth map with custom color scheme
fig, ax = plt.subplots(figsize=(12, 8))
world.plot(column='GDP_per_capita', cmap='YlGnBu', legend=True, ax=ax,
           vmin=threshold_low, vmax=threshold_high)

plt.title('Inequality: GDP per Capita by Country')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```
