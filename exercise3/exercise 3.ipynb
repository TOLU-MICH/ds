# Task 1: Static scatter plot

```python
# static scatter plot
import pandas as pd
from matplotlib import pyplot as plt
import random
#matplotlib notebook

df=pd.read_csv("melb_data.csv")

xValue="Rooms"
yValue="Price"

y=df[yValue]
x=df[xValue]

plt.scatter(x,y, color="green")
plt.xlabel(xValue)
plt.ylabel(yValue)

plt.show()
```

# Task 1: Interacting scatter plot

```python
# interacting scatter plot

# importing the library
import plotly.express as px
import ipywidgets 
import pandas as pd

def slice(no=100):
    df=pd.read_csv("melb_data.csv")[0:no]
    fig = px.scatter(df, x="Rooms", y="Price",  color='Date',symbol="Address")
    fig.show()

ipywidgets.interact(slice, no=[100,500,1000,5000]);
```

# Task 2: static box plot

```python
# Box Plot

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

xValue="Regionname"
yValue="Price"

df = pd.read_csv('melb_data.csv')

ax=df.boxplot(column=yValue,by=xValue,figsize=(20,10))
ax.set_ylabel(yValue)
ax.set_title('Box Plot - Price Vs Rooms')
```

# Task 2: interactive box plot

```python
import plotly.express as px
import pandas as pd

xValue="Regionname"
yValue="Price"

def slice(set_of_data=500):
    df = pd.read_csv('melb_data.csv')[0:set_of_data]
    fig=px.box(df,x=xValue,y=yValue,points="outliers",color="Rooms",title="Box Plot of Price vs Region",hover_data=["YearBuilt"])
    fig.show()

ipywidgets.interact(slice, set_of_data=[500,5000,10000]);
```

# Task 3: static histogram 

```python

import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('melb_data.csv')

xValue="Regionname"
yValue="Price"
prices = df["Price"]
bin=[1000000, 2000000,3000000,4000000,5000000,6000000,7000000,8000000,9000000]

plt.hist(prices,bins=bin,edgecolor='black')
# plt.legend()
plt.title('Property Prices')
plt.xlabel("Prices")
plt.ylabel("Properties")
plt.tight_layout()

plt.show()
```

# Task 3: Interactive histogram 

```python
import pandas as pd
from matplotlib import pyplot as plt
import plotly.express as px

df = pd.read_csv('melb_data.csv')
xValue="Price"

fig=px.histogram(df,x=xValue,height=600,width=800,text_auto=True,color_discrete_sequence=['turquoise'],template="plotly_dark")
fig.update_traces(textposition="outside")
fig.update_layout(title_text="Frequency of the Property "+xValue, title_x=0.5)

fig.show()
```
